```python
import pandas as pd
pd.set_option("display.max_rows", None)
```

```python
import datetime as dt
```

```python
import matplotlib.pyplot as plt
```

```python
Food_Price = pd.read_csv("wfp_food_prices_malawi.csv")
```

```python
Food_Price.head(10)
```

```python
Food_Price['Date'] = pd.to_datetime(Food_Price['Date'])
```

```python
Food_Price['Date'].dt.year == 2019
```

```python
bridged_Food_Price = Food_Price[(Food_Price['Date'].dt.year >= 2014) & (Food_Price['Date'].dt.year <= 2019)]
```

```python
bridged_Food_Price.groupby('Country_Region').mean()
```

```python
Median_Food_Price = bridged_Food_Price.groupby(['Country_Region', 'Date', 'Commodity_Name']).median()
```

```python
Median_Food_Price
```

```python
Median_Food_Price.plot()
```

```python
Median_Food_Price.to_csv('half-baked_Malawi_FP.csv')
```
